## **------------SUPERMARKET INVENTORY MANAGEMENT SYSTEM------------**

# **Read Inventory**



```python
import json
fd = open("records.json",'r')
prod_details = fd.read()
fd.close()

rec_dict = json.loads(prod_details)  # json to dict

rec_dict
```

# **Adding new Product into Inventory**




```python
u_pid = str(input("Enter product id:"))
u_pname = str(input("Enter product name:"))
u_price = int(input("Enter product price:"))
u_quant = int(input("Enter product quantity:"))
u_cat = str(input("Enter product category:"))
u_mfdDate = str(input("Enter manufacturing date:"))
u_expDate = str(input("Enter expiry date:"))

rec_dict[u_pid] = {'name': u_pname, 'price': u_price, 'quantity': u_quant, 'category': u_cat, 'mfdDate': u_mfdDate, 'expDate': u_expDate}

js = json.dumps(rec_dict)   # dict to json

fd = open("records.json",'w')
fd.write(js)
fd.close()
```

```python
rec_dict
```

# **Updating Existing Product in Inventory**

```python
u_pid = str(input("Enter product id:"))

if u_pid in rec_dict:
    u_pname = str(input("Enter product name:"))
    u_price = int(input("Enter product price:"))
    u_quant = int(input("Enter product quantity:"))
    u_cat = str(input("Enter product category:"))
    u_mfdDate = str(input("Enter manufacturing date:"))
    u_expDate = str(input("Enter expiry date:"))
    upd_quant = rec_dict[u_pid]['quantity'] + u_quant   #updating the stock
    rec_dict[u_pid] = {'name': u_pname, 'price': u_price, 'quantity': upd_quant, 'category': u_cat, 'mfdDate': u_mfdDate, 'expDate': u_expDate}
else:
  print("Product Not Found!! Enter the valid Product Id!") 



  
#update records.json file  
js = json.dumps(rec_dict)   # dict to json
fd = open("records.json",'w')
fd.write(js)
fd.close()
```

```python
#update records.json file 
js = json.dumps(rec_dict)   # dict to json

fd = open("records.json",'w')
fd.write(js)
fd.close()
```

# **Deleting the product from Inventory**

```python
u_pid = str(input("Enter product id:"))
if u_pid in rec_dict:
    del rec_dict[u_pid]
else:
  print("Product Not Found!! Enter the valid Product Id!") 
```
